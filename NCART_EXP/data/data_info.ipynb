```python
import os
import pickle
import pandas as pd
import numpy as np
```

```python
clf_data_path_list = ["num_and_cat/classification/",
                "numerical_only/classification/",]
reg_data_path_list = ["num_and_cat/regression/",
                "numerical_only/regression/"]
```

```python

clf_name_list = []
clf_feature_list = []
clf_sample_list = []
clf_target_list = []
objective_list = []

for load_path in  clf_data_path_list:      
    for data_name in os.listdir(load_path):
        data = pickle.load(open(load_path + data_name, 'rb'))
        X = data[0]
        y = data[1]
        num_features = X.shape[1]
        num_samples = X.shape[0]
        clf_name_list.append(data_name)
        clf_feature_list.append(num_features)
        clf_sample_list.append(num_samples)
        clf_target_list.append(str(np.max(y)+1))
        if np.max(y) ==1:
            objective_list.append('binary')
        else:
            objective_list.append('classification')
            
            
        
data_info = {'data name':clf_name_list,
             'num_samples':clf_sample_list,
             'num_feature':clf_feature_list,
             'num_target':clf_target_list,
             'objective': objective_list}
df = pd.DataFrame(data_info)
```

```python
print('binary clf data info:')
df[df['num_target'].str.contains('2')].reset_index(drop=True)
```

```python
print('multi clf data info:')
df[~df['num_target'].str.contains('2')].reset_index(drop=True)
```

```python
reg_name_list = []
reg_feature_list = []
reg_sample_list = []

for load_path in  reg_data_path_list:      
    for data_name in os.listdir(load_path):
        data = pickle.load(open(load_path + data_name, 'rb'))
        X = data[0]
        y = data[1]
        num_features = X.shape[1]
        num_samples = X.shape[0]
        reg_name_list.append(data_name)
        reg_feature_list.append(num_features)
        reg_sample_list.append(num_samples)
        
data_info = {'data name':reg_name_list,
             'num_samples':reg_sample_list,
             'num_feature':reg_feature_list}
df = pd.DataFrame(data_info)
print('reg data info:')
df
```
