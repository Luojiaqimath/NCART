```python
import numpy as np
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import fetch_openml
import os
```

```python

# numerical feature for classification
dir_name = 'numerical_only/classification'

# binary
data_name = 'Higgs'
data = fetch_openml(data_id=44129, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)
        

data_name = 'spambase'
data = fetch_openml(data_id=44, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)


data_name = 'diabetes'
data = fetch_openml(data_id=37, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)


data_name = 'sonar'
data = fetch_openml(data_id=40, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)
        
        
data_name = 'qsar-biodeg' 
data = fetch_openml(data_id=1494, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)


data_name = 'ozone-level-8hr' 
data = fetch_openml(data_id=1487, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)


data_name = 'MagicTelescop'
data = fetch_openml(data_id=44125, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)


data_name = 'jannis'
data = fetch_openml(data_id=44131, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)
        
        
data_name = 'delta_ailerons'     
data = fetch_openml(data_id=803, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)





#  multi-class

data_name = 'EMNIST_Balanced'     
data = fetch_openml(data_id=41039, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)
        

data_name = 'mnist_784'
data = fetch_openml(data_id=554, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)
        
  
data_name = 'semeion'  # 10-class
data = fetch_openml(data_id=1501, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)
        

data_name = 'one-hundred-plants-margin'  # 100-class
data = fetch_openml(data_id=1491, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f) 
        

data_name = 'optdigits'  
data = fetch_openml(data_id=28, parser='liac-arff')  
X = data.data.values.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y), f)  
```

```python
# num_and_cat feature for classification
dir_name = 'num_and_cat/classification'

# binary
data_name = 'credit-g'
data = fetch_openml(data_id=31, parser='liac-arff')  
X = data.data.values
categorical_idx = [0, 2, 3, 5, 6, 8, 9, 11, 13, 14, 16, 18, 19]
for i in categorical_idx:
    le = LabelEncoder()
    X[:, i] = le.fit_transform(X[:, i])
X = X.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y, categorical_idx), f)


data_name = 'scene'
data = fetch_openml(data_id=312, parser='liac-arff')  
X = data.data.values  # last five is cat idx
categorical_idx = [294, 295, 296, 297, 298]
for i in categorical_idx:
    le = LabelEncoder()
    X[:, i] = le.fit_transform(X[:, i])
X = X.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y, categorical_idx), f)
        

data_name = 'road-safety'
data = fetch_openml(data_id=44161, parser='liac-arff')  
X = data.data.values
categorical_idx = [6, 22, 25]
for i in categorical_idx:
    le = LabelEncoder()
    X[:, i] = le.fit_transform(X[:, i])
X = X.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y, categorical_idx), f)
        
        
        
        
        
# multi-class
data_name = 'vowel'
data = fetch_openml(data_id=307, parser='liac-arff')  
X = data.data.values
categorical_idx = [0, 1]
for i in categorical_idx:
    le = LabelEncoder()
    X[:, i] = le.fit_transform(X[:, i])
X = X.astype(np.float32)
le = LabelEncoder()
y = le.fit_transform(data.target.values)
if not os.path.exists(dir_name):
    os.makedirs(dir_name)
with open("{}/{}".format(dir_name, data_name), "wb") as f:
        pickle.dump((X, y, categorical_idx), f)
```
